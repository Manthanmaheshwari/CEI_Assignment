# Single Agent Pipeline Project

## Problem Statement
Build a Single-Agent Smart Assistant that:
- Understands user queries
- Routes tasks based on intent
- Uses tools when required
- Returns structured JSON output

### The agent should handle:
- Math queries: Calculator Tool
- Keyword extraction: Keyword Tool
- General queries: Direct response

---
### Implementation Requirements
- Agent logic
- Conditional routing
- Tool integration
- Basic error handling

````python
def calculator(expression: str) -> str:
    """Evaluate a mathematical expression safely."""
    try:
        # Basic sanitization of the input string
        allowed_chars = "0123456789+-*/(). "
        if all(c in allowed_chars for c in expression):
            return str(eval(expression))
        else:
            return "Invalid characters in expression"
    except Exception as e:
        return f"Error in calculation: {str(e)}"
````

````python
def extract_keywords(text: str) -> list:
    """Extract unique words longer than 4 characters from the text."""
    try:
        words = text.split()
        # Filter for words longer than 4 characters and remove duplicates
        keywords = sorted(list(set([w.lower().strip(".,!?;:") for w in words if len(w) > 4])))
        return keywords[:5]
    except Exception:
        return []
````

## 🤖 Implement Agent Logic Below

👉 Use conditional routing:
- If query contains "calculate" → use calculator
- If query contains "keywords" → use keyword extractor
- Else → general response

````python
def agent(query: str):
    """
    Primary agent function that routes the user query to the appropriate tool
    based on keywords and returns a structured JSON response.
    """
    query_lower = query.lower()

    try:
        # Routing for calculation tasks
        if "calculate" in query_lower:
            # Extract the expression by removing the command word
            expression = query_lower.replace("calculate", "").strip()
            result = calculator(expression)
            return {
                "type": "calculation",
                "result": result
            }

        # Routing for keyword extraction tasks
        elif "keywords" in query_lower:
            # Extract the text following the command word
            text_content = query_lower.split("keywords from", 1)[-1].strip() if "keywords from" in query_lower else query_lower.replace("keywords", "").strip()
            result = extract_keywords(text_content)
            return {
                "type": "keywords",
                "result": result
            }

        # Default routing for general queries
        else:
            return {
                "type": "general",
                "result": f"I received your query: '{query}'. How can I help you further?"
            }

    except Exception as e:
        return {
            "type": "error",
            "result": str(e)
        }
````

## 📦 Expected Output Format

```
{
  "type": "calculation / keywords / general / error",
  "result": ...
}
```

````python
# Define the test cases to verify the agent's routing and tool usage
queries = [
    "Calculate 20 + 5",
    "Extract keywords from Artificial Intelligence is transforming industries",
    "What is machine learning?"
]

# Execute each query through the agent and display the JSON result
for query in queries:
    print(f"Query: {query}")
    response = agent(query)
    print(f"Response: {response}")
    print("-" * 50)
````

````python
# 🎯 Interactive Mode

while True:
    user_input = input("Enter query (type 'exit' to stop): ")
    if user_input.lower() == "exit":
        break
    print("Response:", agent(user_input))
````
